### Introduction 

Using basic statistics concepts, such a bar graph and time series, this project explores two dataset containing information about crude oil exports and production by the Energy Commission of Ghana from 2000 to 2019.

### Dataset Description

The datasets were downloaded from the Ghana Open Data Initiative website via this link: (https://data.gov.gh/dataset/crude-oil-2000-2019). The datasets were made available by the Energy Commission of Ghana and can be accessed by the general public. The datasets have two columns: Year and Crude Oil Exports/Crude Oil Production with 20 entries. 


### Importing and loading packages and datasets

```python
### load crude oil dataset

df1 = pd.read_csv('Crude Oil Export (Million barrels) 2002 - 2019.csv', index_col=False)
df1.head()
```

```python
## load crude oil production dataset
df2 = pd.read_csv('Crude Oil Production (million barrels) 2002 - 2019.csv', index_col=False)
df2.head()
```

## Data Cleaning

## Issues

### Issue 1: Rename columns for both dataframes to match data values

```python
### Rename column header for df1

df1_clean = df1.rename(columns={"Crude Oil Export (Million barrels)": "Year", "Unnamed: 1": "Crude Export (Million Barrels)"})
df1_clean
```

```python
## Rename column header for df2 

df2_clean = df2.rename(columns={"Crude Oil Production (million barrels)": "Year", "Unnamed: 1": "Crude Oil Production (Million Barrels)"})
df2_clean
```

### Issue 2: Delete rows that are not needed in the dataframe

```python
# Delete error rows 
df1_clean.drop(labels=[0,1], inplace=True)
df1_clean
```

```python
df2_clean.drop(labels=[0,1], inplace=True)
df2_clean
```

## Issue 3: Convert year column to integer

```python
## convert year column to integer

df1_clean['Year'].astype(int)
df1_clean
```

```python
df2_clean['Year'].astype(int)
df2_clean
```

## Issue 4: Convert crude oil export and production column to float

```python
## Convert crude oil export and production to integer

df1_clean['Crude Export (Million Barrels)'].astype(float)
df1_clean
```

```python
df2_clean['Crude Oil Production (Million Barrels)'].astype(float)
df2_clean
```

## Univariate Exploration 

```python
## A bar graph to display the Crude Oil Export (Million barrels) between from 2002 to 2019.

x = df1_clean['Year']
y = df1_clean['Crude Export (Million Barrels)']
plt.bar(x, y, color ='maroon', width = 0.6)
 
plt.xlabel("Year")
plt.ylabel("Crude Export (Million Barrels)")
plt.title("Graph of Crude Oil Export (Million barrels) for the Years 2002 to 2019.")
plt.rcParams["figure.figsize"] = 15,5
plt.show()
```

```python
x2 = df2_clean['Year']
y2 = df2_clean['Crude Oil Production (Million Barrels)']
plt.bar(x2, y2, color ='maroon', width = 0.6)
 
plt.xlabel("Year")
plt.ylabel("Crude Oil Production (Million Barrels)")
plt.title("Graph of Crude Oil Production (Million Barrels) for the Years 2002 to 2019.")
plt.rcParams["figure.figsize"] = 15,10
plt.show()
```

```python
plt.plot(df1_clean.Year, df1_clean['Crude Export (Million Barrels)'], label='Crude Oil Export', 
         linewidth=3)
plt.plot(df2_clean.Year, df2_clean['Crude Oil Production (Million Barrels)'], label='Crude Oil Production', 
         linewidth=3)
plt.rcParams["figure.figsize"] = 15,10
plt.legend()
```

## Conclusion 

I analysed the Crude Oil Export and Production datasets in this project. From the bar graphs, both crude oil exports and production, in millions of barrels continued to increase throughout the years. However, the time series analysis showed that from 2002 to 2009, Ghana was producing and exporting the same amount of crude oil.
